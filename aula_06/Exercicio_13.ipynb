```python
# Objetivo: Realizar um left join para exibir todos os livros, independentemente de terem sido emprestados.

# Instruções:

# 1. Carregue novamente os arquivos.
# 2. Realize um left join entre as tabelas livros e emprestimos pela coluna id_livro.
# 3. Exiba os livros que não foram emprestados (os que têm valores NaN nas colunas relacionadas aos empréstimos).
```

```python
import pandas as pd

# 1 Carregando os arquivos.
df_livros = pd.read_csv('livros.csv')
df_emprestimos = pd.read_csv('emprestimos.csv')
```

```python
# 2 Realizando o left entre as tabelas pelo id_livro.
result = pd.merge(df_livros, df_emprestimos, on='id_livro', how='left')
result
```

```python
# Exibindo livros que nao foram emprestados.
nao_emprestados = result[result['data_emprestimo'].isna()]
nao_emprestados
```
