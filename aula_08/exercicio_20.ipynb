energy-consumption
Faça um join entre os dados de consumo de energia e emissões de CO2 por país/continente.

```python
import pandas as pd
import matplotlib.pyplot as plt

df_energia = pd.read_csv("dados_energia_simulados_ptbr.csv")
#print(df_energia["Continente"].unique())
df_energia
```

### Crie um Boxplot que compare o consumo de energias renováveis e fósseis por continente.

```python
# Grafico de box.
for i in df_energia["Continente"].unique():
    df_energia_regiao = df_energia[df_energia['Continente'] == i]
    fig,ax = plt.subplots()
    ax.boxplot([df_energia_regiao['Consumo_Energia_Renovável (MWh)']], ['Consumo_Energia_Fóssil (MWh)'])
    ax.set_title(f"Gráfico de box do continente {i}")
    plt.show()
```

Use gráficos de linha para mostrar a evolução do consumo de energias renováveis e fósseis e as emissões de CO2 ao longo do tempo.
