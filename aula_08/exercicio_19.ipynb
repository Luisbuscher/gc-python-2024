### Crie gráficos para visualizar as diferenças entre as espécies de pinguins com base nas características físicas. Use gráficos de dispersão, boxplot e histogramas para mostrar as distribuições e comparações entre as variáveis.

```python
# Grafico de dispersao.
plt.scatter(df_pinguim['body_mass_g'], df_pinguim['bill_length_mm'], color = 'purple')
plt.title('Grafico de dispersao')
plt.xlabel('body_mass_g')
plt.ylabel('bill_length_mm')
plt.grid()
plt.show()
```

```python
# Grafico de histograma.
plt.hist(df_pinguim['body_mass_g'], color = 'purple')
plt.title('Grafico de histograma')
plt.xlabel('body_mass_g')
plt.ylabel('bill_length_mm')
plt.grid()
plt.show()
```

```python
# Grafico de box.
fig,ax = plt.subplots()
ax.boxplot([df_pinguim['body_mass_g']], ['bill_length_mm'])
ax.set_title("Gráfico de box")
plt.show()
```
